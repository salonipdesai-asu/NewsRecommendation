```python
import pandas as pd
from flask import Flask, request, jsonify
from flask_cors import CORS
```

```python
app = Flask(__name__)
CORS(app)
```

```python
def read():
    data = pd.read_csv("C:/Users/JAS/Desktop/Dataset/newsrecomm2new.csv")
    return data
```

```python
@app.route('/',methods=['POST'])
def rate():
    #read csv and process dataframe
    data = read()
    if 'Unnamed: 0' in data.columns:
        data = data.drop('Unnamed: 0', axis=1)
    data.set_index('url',inplace=True)
    #get data from post request
    userData = request.get_json(force=True)
    if(userData['url'] in data.index):
        data.loc[userData['url'],userData['email']]=userData['rating']
    else:
        data1 = pd.DataFrame({"url":[userData['url']],userData['email']:[userData['rating']]})
        data1.set_index("url", inplace=True)
        data = pd.concat([data,data1])
    data.to_csv("C:/Users/JAS/Desktop/Dataset/newsrecomm2new.csv")
    return jsonify(str(data.loc[userData['url'],userData['email']]))
```

```python
if __name__ == '__main__':
    app.run(port=5002, debug=False)
```
